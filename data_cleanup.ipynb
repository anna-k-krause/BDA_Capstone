### Data Cleaning

```python
# Find the columns where each value is null
empty_cols = [col for col in df.columns if df[col].isnull().all()]
# Drop these columns from the dataframe
df.drop(empty_cols,
        axis=1,
        inplace=True)
```

```python
# Get Rid of Columns that have fewer than 250 values (1% of all reviews count)
small_cols = [col for col in df.columns if df[col].count() < 250]
df.drop(small_cols,
        axis=1,
        inplace=True)
```

```python
# remove columns listed as inactive
df = df[df.columns.drop(list(df.filter(regex='Inactive')))]
```

```python
# remove columns listed as inactive
df = df[df.columns.drop(list(df.filter(regex='inactive')))]
```

```python
# dropping columns deemed useless (conatined the same response for all reviews)
df_clean = df.drop(['Brand',
                   'Marriott Survey Type',
                   'Marsha Code',
                   'None'],
                   axis=1)
```

```python
df_clean['Survey language'].value_counts()
```

```python
# removing all non-English reviews
lang = ['English','English (United States)','English (UK)']
df_english = df_clean[df_clean['Survey language'].isin(lang)]
```

```python
df_english.head()
```

```python
df_english.info()
```

```python
df_english.to_csv('marriott_clean.csv',index=False)
```
